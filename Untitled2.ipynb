# 3. Montagem do gráfico de dispersão e cálculo da correlação e covariância.

Construíremos o gráfico de forma que o eixo das abscissas represente as horas estudadas pelos alunos e o eixo das ordenadas, as notas  do exame.


```python
plt.scatter(df['Hours_Studied'], df['Exam_Score'])
plt.xlabel('Horas estudadas')
plt.ylabel('Notas do exame')
plt.title('Relação entre horas estudadas e notas do exame')
plt.show()
```

Agora, iremos calcular a correlção entre as horas estudadas e as notas do exame.

```python
df['Hours_Studied'].corr(df['Exam_Score'])
```

Como o valor está entre 0.9 e 1.0, podemos afirmar que temos uma correlação forte entre as horas estudadas e as notas do exame. Agora, verificaremos a covariância.

```python
covariancia = df['Hours_Studied'].cov(df['Exam_Score'])
print(f"Covariância: {covariancia:.2f}")
```

# 4. Cálculo da reta de regressão.

Para calcular esta reta, utilizarei o método de regressão linear, utilizando a seguinte fórmla:

```python
# Médias
media_x = df['Hours_Studied'].mean()
media_y = df['Exam_Score'].mean()

# Varância de x
var_x = df['Hours_Studied'].var()

# Varância de y
var_y = df['Exam_Score'].var()

# Covarância
cov_xy = df['Hours_Studied'].cov(df['Exam_Score'])

# Coeficientes
beta_1 = cov_xy / var_x
beta_0 = media_y - beta_1 * media_x

print(f"Equação da Reta")
print(f"Nota = {beta_0:.2f} + {beta_1:.2f} x Notas")
```

Interpretação:

Com base no que foi calculado acima, o valor beta_0, o intercepto, é uma possível nota para um aluno que estudou 0 horas. Para beta_1, a inclinação, temos que cada hora de estudo que um aluno realiza, sua nota será aumenta em 5.26 pontos percentuais.

# 5. Previsão de notas.

Nesta seção, utilizaremos tempo de estudo específico a fim de retornar uma estimativa para a nota que o aluno poderá tirar com base em meus cálculos.

```python
# Previsão de nota de um aluno que estudou por 7 horas.
horas = 7
nota_prevista = beta_0 + beta_1 * horas
print(f"Aluno que estuda {horas} horas: nota prevista = {nota_prevista:.1f}%")
```

# 6. Construção do gráfico da reta de regressão.

```python
# Pontos reais.
plt.scatter(df['Hours_Studied'], df['Exam_Score'], color = 'blue', label = 'Dados reais')

# Reta de regressão.
x_vals = np.array([0,10]) # de 0 a 10 horas
y_vals = beta_0 + beta_1 * x_vals
plt.plot(x_vals, y_vals, color = 'red', label = 'Reta de regressão')

plt.xlabel('Horas de estudo')
plt.ylabel('Nota')
plt.title('Regressão Linear - Horas vs Notas')
plt.legend()
plt.show()
```

# 7. Cáculo dos erros e construção do histograma.

```python
# Calcular nota prevista para cada aluno.
df['Nota_Prevista'] = beta_0 + beta_1 * df['Hours_Studied']
df['Erro'] = df['Exam_Score'] - df['Nota_Prevista']

# Mostrar as primeiras linhas
print(df.head())

# Média dos erros
print(f"Média dos Erros: {df['Erro'].mean():.10f}")
```

Agora, com base nos cálculos realizados acima, contruírei o histograma dos erros.

```python
plt.hist(df['Erro'], bins=5, edgecolor='black')
plt.xlabel('Erro (pontos percentuais)')
plt.ylabel('Frequência')
plt.title('Distribuição dos Erros')
plt.show()
```
